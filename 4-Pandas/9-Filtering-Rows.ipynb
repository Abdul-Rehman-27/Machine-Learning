## Filtering Rows

```python
# Filtering Rows Based On:
'''
1-Single Condition 
2-Multiple Condition 
3-Used Boolean Indexing for filtering Rows
'''
```

```python
import pandas as pd

data = {
    "Name": ["Ahmed", "Fatima", "Hassan", "Ayesha", "Bilal", "Zainab", "Umar", "Maryam"],
    "Age": [28, 34, 22, 30, 29, 40, 25, 32],
    "Salary": [50000, 60000, 45000, 52000, 49000, 70000, 48000, 50000],
    "Performance Score": [85, 90, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(f'Sample Data Frame\n {df}')
```

```python
# Single Condition 
high_salary = df[df['Salary']>50000]
print('Employe Salary With Greater than 50000')
print(high_salary)
```

```python
# Multiple Condition 
filtered = df[(df['Salary'] > 50000) & (df['Age'] > 30)]
print('Employees with Salary > 50k and Age > 30')
print(filtered)
```
