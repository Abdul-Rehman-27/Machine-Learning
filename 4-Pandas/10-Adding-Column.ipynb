## Adding COlumn In DF

```python
import pandas as pd

data = {
    "Name": ["Ahmed", "Fatima", "Hassan", "Ayesha", "Bilal", "Zainab", "Umar", "Maryam"],
    "Age": [28, 34, 22, 30, 29, 40, 25, 32],
    "Salary": [50000, 60000, 45000, 52000, 49000, 70000, 48000, 50000],
    "Performance Score": [85, 90, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(f'Sample Data Frame\n {df}')
```

```python
# Adding Column Via Asssignment
df['Bonus'] = df['Salary']*0.1
print('New data frame with new column')
print(df)
```

```python
# Adding Column via insert method at specific position while in above we can't do in above 
#df.inser(loc,'column-name',some-data)
df.insert(0,'Employe-Id',[1,2,3,4,5,6,7,8])
print('Adding new column at precise index via insert method')
print(df)
```
