## Use of info() Method

```python
# Problem
''' 
1-No of rows and columns
2-What type of
3-Missing Data
4- Memory Usage
'''
#Info Methos Tells:
'''
1-No of rows and COlumns
2-Columns names
3-Data Types
4-Non Null counts
5-Memory Uage of DataFrame
'''
```

```python
import pandas as pd

data = {
    "Name": ["Ahmed", "Fatima", "Hassan", "Ayesha", "Bilal", "Zainab", "Umar", "Maryam"],
    "Age": [28, 34, 22, 30, 29, 40, 25, 32],
    "Salary": [50000, 60000, 45000, 52000, 49000, 70000, 48000, 50000],
    "Performance Score": [85, 90, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(f'Sample Data Frame\n {df}')
```

```python
print('Display Info of Data Set')
print(df.info())
```
