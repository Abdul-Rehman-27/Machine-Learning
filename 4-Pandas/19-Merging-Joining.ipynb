# Merging and joining 

```python
import pandas as pd

df_customers = pd.DataFrame({
    'Customer ID': [1, 2, 3],
    'Name': ['Ali', 'Ahmed', 'Farhan']
})

df_orders = pd.DataFrame({
    'Order ID': [1, 2, 4],
    'Customer ID': [1, 2, 4],
    'Amount': [250, 450, 350]
})

print("Customers DataFrame:")
print(df_customers)

print("Orders DataFrame:")
print(df_orders)
```

```python
#pd.merge(df1, df2, on='Column-name', how='Type of join')
df_merged = pd.merge(df_customers, df_orders, on='Customer ID', how='inner')
print('Inner Join')
print(df_merged)
```

```python
#pd.merge(df1, df2, on='Column-name', how='Type of join')
df_merged = pd.merge(df_customers, df_orders, on='Customer ID', how='outer')
print('Outer Join')
print(df_merged)
```

```python
#pd.merge(df1, df2, on='Column-name', how='Type of join')
df_merged = pd.merge(df_customers, df_orders, on='Customer ID', how='left')
print('Left Join')
print(df_merged)
```

```python
#pd.merge(df1, df2, on='Column-name', how='Type of join')
df_merged = pd.merge(df_customers, df_orders, on='Customer ID', how='right')
print('Right Join')
print(df_merged)
```

```python
#pd.merge(df1, df2, how='Type of join')
df_merged = pd.merge(df_customers, df_orders, how='cross')
print('Cross Join')
print(df_merged)
```
