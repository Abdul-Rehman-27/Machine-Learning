## Detecting Missing Values

```python
# Missing Data is what?
'''
1-NaN (not a number)
2-None (for object data types)
'''
# How to Detect Missing Data
'''
1-Before Handling the data u need to know where data is missing for this we use:
    isnull() return:
      True: mean NaN (Value is Missing)
      False: if false mean value is present
'''
```

```python
import pandas as pd

data = {
    "Name": ["Ahmed", None, "Hassan", "Ayesha", "Bilal", "Zainab", "Umar", "Maryam"],
    "Age": [28, None, 22, 30, 29, 40, 25, 32],
    "Salary": [50000, None, 45000, 52000, 49000, 70000, 48000, 50000],
    "Performance Score": [85, None, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(f'Sample Data Frame\n {df}')
```

```python
print(df.isnull())
```

```python
# if u also want to check how much value is missing
print(df.isnull().sum())
```

```python
#Key Points:
'''
Check the percetnage of missing data
- df.isnull().mean() * 100
'''
```
