## Rows Filtering 

```python
# With single Condition 
view_dataFrame.loc[(view_dataFrame["A"]<0.3)] #Returns all rows of dataFrame where the value in column "A" is less than 0.3.
```

```python
# With multiple Condition 
view_dataFrame.loc[(view_dataFrame["A"]<0.5) & ((view_dataFrame["C"]>0.2))] #Returns all rows of dataFrame where the value in column "A" is less than 0.3 and Value in Column C is > 0.1
```

## New DataFrame

```python
new_dataFrame = pd.DataFrame({
    "name": ['Alfred', 'Batman', 'Catwoman'],
    "toy": [np.nan, 'Batmobile', 'Bullwhip'],
    "born": [pd.NaT, pd.Timestamp("1940-04-25"), pd.NaT]
})
```

```python
new_dataFrame
```

## dropna Function 

```python
new_dataFrame.dropna()
```

## Use of how

```python
new_dataFrame.dropna(how="all") # delete row or columns unless all values are not na
```

```python
new_dataFrame.dropna(how="any")   # If any value in row is na delete that row
```

## New DataFrame_2 

```python
new_dataFrame_2 = pd.DataFrame({
    "name": ['Alfred', 'Batman', 'Catwoman'],
    "toy": [np.nan, np.nan, np.nan],
    "born": [pd.NaT, pd.Timestamp("1940-04-25"), pd.NaT]
})
```

```python
new_dataFrame_2.dropna(how="all", axis=1)
```

## New DataFrame_3

```python
new_dataFrame_3 = pd.DataFrame({
    "name": ['Alfred', 'Batman', 'Alfred'],
    "toy": [np.nan, 'Batmobile', 'Bullwhip'],
    "born": [pd.NaT, pd.Timestamp("1940-04-25"), pd.NaT]
})
```

```python
new_dataFrame_3["name"].value_counts(dropna=False)    # Also keep na Value and count
```

```python
new_dataFrame_3["toy"].value_counts(dropna=True)  #Remove na and count
```
