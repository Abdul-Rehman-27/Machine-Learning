## Use Of Head and Tail

```python
import pandas as pd

data = {
    "Name": ["Ahmed", "Fatima", "Hassan", "Ayesha", "Bilal", "Zainab", "Umar", "Maryam"],
    "Age": [28, 34, 22, 30, 29, 40, 25, 32],
    "Salary": [50000, 60000, 45000, 52000, 49000, 70000, 48000, 50000],
    "Performance Score": [85, 90, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(f'Sample Data Frame\n {df}')
```

```python
# Use of Head
head = df.head(2)
print('Display First 2 rows of Df')
print(head)
```

```python
# Use of tail
tail = df.tail(3)
print('Display Last 3 Rows of Df')
print(tail)
```
