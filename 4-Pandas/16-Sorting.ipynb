# Sorting

```python
import pandas as pd

data = {
    "Name": ["Ahmed", "Fatima", "Hassan", "Ayesha", "Bilal", "Zainab", "Umar", "Maryam"],
    "Age": [28, 34, 22, 30, 29, 40, 25, 32],
    "Salary": [50000, 60000, 45000, 52000, 49000, 70000, 48000, 50000],
    "Performance Score": [85, 90, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(f'Sample Data Frame\n {df}')
```

```python
#Sorting single columns
df.sort_values(by='Age', ascending=True, inplace=True)
print('Sorted Age by ascending')
print(df)
```

```python
#Sorting Multiple Columns
df.sort_values(by=['Age','Salary'], ascending=[True,True], inplace=True)
print('Sorted Age by ascending')
print(df)
```
