```python
from PIL import Image

# Load your image
image_path = "primary.png"  # Replace with your file name
original_image = Image.open(image_path)

# Resize to 1280x769
resized_image = original_image.resize((1280, 769))

# Estimate area of "Click Here" button (bottom-right corner)
width, height = resized_image.size
cropped_area = resized_image.crop((0, 0, width - 250, height - 130))  # Adjust if needed

# Paste the cropped area back on to hide the button
final_image = resized_image.copy()
final_image.paste(cropped_area, (0, 0))

# Save with 72 DPI
final_image.save("AI_Services_1280x769_no_click_here.png", dpi=(72, 72))
print("Image saved successfully!")
```
