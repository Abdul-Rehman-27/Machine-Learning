```python
import pandas as pd

df = pd.DataFrame({
    "name": ['Alfred', 'Batman', 'Alfred', 'Batman'],
    "toy": ['None', 'Batmobile', 'None', 'Batmobile'],
    "city": ['Lahore', 'Gotham', 'Lahore', 'Gotham']
})

print("Original DataFrame")
print(df)
```

```python
remove_duplicates = df.drop_duplicates(subset=["name"])
print('Df after removing duplicates')
print(remove_duplicates)
```

```python
remove_duplicates = df.drop_duplicates(subset=["name"], keep='first')
print('Df after removing duplicates (keep first):')
print(remove_duplicates)
```

```python
remove_duplicates = df.drop_duplicates(subset=["name"], keep='last')
print('Df after removing duplicates (keep last):')
print(remove_duplicates)
```

```python
remove_duplicates = df.drop_duplicates(subset=["name"], keep=False)
print('Df after removing duplicates (keep drop):')
print(remove_duplicates)
```
