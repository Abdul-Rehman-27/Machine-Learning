## Concatinating Data Frame

```python
# What is Concatination
'''
Combining Data frames:
1-vertically (row wise)
2-horizontally (column wise)

Syntax:
pd.concate([df1,df2], axis=0, ignore_index=True)  
ignore_index=True mean reset index after combining
'''
```

```python
import pandas as pd

df_reg1 = pd.DataFrame({
  'Customer Id':[1,2],
  'Name':['Bilal','Haroon']
})

df_reg2 = pd.DataFrame({
  'Customer Id':[3,4],
  'Name':['babu','ouk']
})

df_concate = pd.concat([df_reg1, df_reg2], axis=0, ignore_index=True)
print(f'Vertically Concate Df')
print(df_concate)
```

```python
#Horizontally Concate
df_concate = pd.concat([df_reg1, df_reg2], axis=1, ignore_index=True)
print(f'Horizontally Concate Df')
print(df_concate)
```
