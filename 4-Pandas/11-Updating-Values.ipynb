## Updating Value in Df

```python
import pandas as pd

data = {
    "Name": ["Ahmed", "Fatima", "Hassan", "Ayesha", "Bilal", "Zainab", "Umar", "Maryam"],
    "Age": [28, 34, 22, 30, 29, 40, 25, 32],
    "Salary": [50000, 60000, 45000, 52000, 49000, 70000, 48000, 50000],
    "Performance Score": [85, 90, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(f'Sample Data Frame\n {df}')
```

## Use Of Loc (loc is label-based (uses column/row names))

```python
#.loc[]
#df.loc[row_index, column_name]
df.loc[0,'Salary'] = 55000
print('Updated Salary of Ahmed')
print(df)
```

```python
#Updating whole Columns
df['Salary'] = df['Salary']*1.05
print('Df after increasing salary by 5%')
print(df)
```
