## Interpolation 

```python
# What is interpolation 
'''
Is a tecnhique by which u fill estimated values in place of missing values
For  example
10
20
NaN
40
50   

Advantages of interpolation:
1-Preserve data integrity(instead of removing rows or columns it will place them with arbitrary values)
2-Smooth Trends(In time series data)
3-Avoid Data Loss

Method:
interpolate()
'''

# When to use interpolaion
'''
1-For time Series data
2-Numeric Data with trends
3-Avoid droppping rows
'''
```

```python
import pandas as pd

data = {
    "Name": ["Ahmed", None, "Hassan", "Ayesha", "Bilal", "Zainab", "Umar", "Maryam"],
    "Age": [28, None, 22, 30, 29, 40, 25, 32],
    "Salary": [50000, None, 45000, 52000, 49000, 70000, 48000, 50000],
    "Performance Score": [85, None, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(f'Sample Data Frame\n {df}')
```

```python
#Method:- Linear, polynomial, Time
df.interpolate(method='linear', axis=0, inplace=True)
print('Df after Interpolation')
print(df)
```
