## Creating Basic Data Frame

```python
dict_1 = {
    "Name" : ["Abdul Rehman", "Haroon", "Zeeshan"],
    "CGPA" : [3.80,3.01,2.75],
    "City" : ["Hazro", "Malak Mala", "Attock"]
}

dataFrame = pd.DataFrame(dict_1)
dataFrame
```
