## Remvoing Columns in df

```python
import pandas as pd

data = {
    "Name": ["Ahmed", "Fatima", "Hassan", "Ayesha", "Bilal", "Zainab", "Umar", "Maryam"],
    "Age": [28, 34, 22, 30, 29, 40, 25, 32],
    "Salary": [50000, 60000, 45000, 52000, 49000, 70000, 48000, 50000],
    "Performance Score": [85, 90, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(f'Sample Data Frame\n {df}')
```

```python
#Drop Method
#df.drop(coulmns = ['columns-name'],inplace=True)
df.drop(columns=['Performance Score'],inplace=True) # inplace=True in Pandas means change the original data directly instead of making a new copy)
print('After Remvoing Performance Columns')
print(df)
```

```python
# Deleting Multiple Columns
df.drop(columns=['Age','Salary'],inplace=True)
print('After Remvoing Age and Salary Columns')
print(df)
```
