## Handling Missing Values

```python
# u have two ways to overcome this prblem
'''
1-Remove that Values
2 or fill that values
'''
#Suppose
'''
1-If the values dont affect your dataset = remove that values by using
      df.dropna(axis=0, inplace=True)
2-If the Values affect your dataset = fill that value by using
      df.fillna(value, inplace=True)
'''
```

```python
import pandas as pd

data = {
    "Name": ["Ahmed", None, "Hassan", "Ayesha", "Bilal", "Zainab", "Umar", "Maryam"],
    "Age": [28, None, 22, 30, 29, 40, 25, 32],
    "Salary": [50000, None, 45000, 52000, 49000, 70000, 48000, 50000],
    "Performance Score": [85, None, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(f'Sample Data Frame\n {df}')
```

```python
# Remvoing missing values
# df.dropna(inplace=True)
# print('Df after removing missing values')
# print(df)
```

```python
# Filling Missing value with default values
df.fillna(27, inplace=True)
print('Df after filling missing values')
print(df)
```

```python
# Filling missing with calcualted values 
df['Age'].fillna(df['Age'].mean(), inplace=True)  #This will give warning
print('Df after filling missing value with calculated values:')
print(df)
```

```python
# Filling missing with calcualted values without facing warning
df['Age'] = df['Age'].fillna(df['Age'].mean())
print('Df after filling missing value with calculated values without facing warning:')
print(df)
```
