## Group by

```python
import pandas as pd

data = {
    'Department': ['IT', 'HR', 'IT', 'HR', 'Finance'],
    'Salary': [50000, 40000, 60000, 45000, 70000],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male']
}

df = pd.DataFrame(data)
print(f'Sample Data Frame\n{df}')
```

```python
#Group by on single column
grouped = df.groupby('Department')['Salary'].sum()
print(grouped)
```

```python
#Group by on single column
grouped = df.groupby(['Department','Gender'])['Salary'].sum()
print(grouped)
```
