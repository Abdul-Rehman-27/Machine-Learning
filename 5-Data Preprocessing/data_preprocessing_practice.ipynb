```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
df = pd.read_csv('data_preprocessing_practice.csv')
print('Original DataFrame')
print(df)
```

# Handling Missing Values

```python
#Checking Percetnage of Missing Values
print("Percentage of Missing Values in each column:")
df.isnull().mean()*100
```

```python
#Detecting Total Missing Values in each column 
print("Detecting Total Missing Values in each column:")
df.isnull().sum()
```

```python
##Droping Missing Values
# df.dropna(inplace=True)
# print("DataFrame after dropping missing values:")
# print(df)
```

```python
df['Gender'] = df['Gender'].fillna('Unknown')
print('DataFrame after filling missing value in gender with Unknown') 
print(df)
      
```

```python
#Filling Missing Values in age with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
print('DataFrame after filling missing value in age with mean')
print(df)
```

```python
#Filling Missing Values in income with mean
df['Income'] = df['Income'].fillna(df['Income'].mean())
print('DataFrame after filling missing value in income with mean')
print(df)
```

```python
#Checking Percetnage of Missing Values
print("Percentage of Missing Values in each column:")
df.isnull().mean()*100
```

# Encoding Categorical Data

```python
# Label Encoding
df_label = df.copy()
le = LabelEncoder()

df_label['Gender_Encdoed'] = le.fit_transform(df_label['Gender'])
print("\nDataFrame after Label Encoding:")
print(df_label.head())
```
