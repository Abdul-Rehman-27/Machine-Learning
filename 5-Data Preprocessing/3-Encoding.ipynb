```python
#Why need of encoding?
'''Machine learning algorithms work with numbers, not strings.
For example
Car Brand: Toyota, Honda, Ford
Fuel Type: Petrol, Diesel, Electric

Ml agortihm got confused ans said
Toyota? Honda? Ford? What is this?

So we need to convert these strings into numbers.
Encoding helps us convert categorical data into numerical data.
'''

#Types of Encoding
'''1.Label Encoding: Assigns a unique integer to each category.
For example:
  Car Brand: Toyota -> 0, Honda -> 1, Ford -> 2

2.One-Hot Encoding: Creates binary columns for each category. Each category is represented by a binary vector.
For example:
  Toyota -> [1, 0, 0], 
  Honda -> [0, 1, 0], 
  Ford -> [0, 0, 1]
'''
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

df = pd.read_csv('sample_data.csv')
print("Original DataFrame:")
print(df)

# Label Encoding
df_label = df.copy()
le = LabelEncoder()

df_label['Gender_Encdoed'] = le.fit_transform(df_label['Gender'])
df_label['Passed_Encoded'] = le.fit_transform(df_label['Passed'])
print("\nDataFrame after Label Encoding:")
print(df_label.head())

```

```python
#One hot encoding
df = pd.read_csv('sample_pakistani_cities.csv')
print("\nOriginal DataFrame")
print(df)

# One-Hot Encoding
df_one_hot = pd.get_dummies(df, columns=['City'], dtype='int')
print("\nDataFrame after One-Hot Encoding:")
print(df_one_hot.head())
```
