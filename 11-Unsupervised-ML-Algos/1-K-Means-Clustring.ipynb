### Simple K-Mean-Clustring 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# ---- Sample 2D points ----
points = np.array([
    [1, 2], [1.5, 1.8], [2, 1.5],
    [8, 8], [8.5, 9], [9, 8.5],
])

# ---- K-Means ----
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(points)

labels = kmeans.labels_           # cluster number of each point
centroids = kmeans.cluster_centers_  # cluster centers

print("Labels:", labels)
print("Centroids:\n", centroids)

# ---- Visualization ----
plt.figure(figsize=(6, 6))

# scatter plot of points colored by cluster
plt.scatter(points[:, 0], points[:, 1], c=labels, s=100)

# plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], 
            marker='X', s=300, c='red', label='Centroids')

plt.title("K-Means Clustering on Simple 2D Points")
plt.xlabel("X coordinate")
plt.ylabel("Y coordinate")
plt.legend()
plt.grid(True)
plt.show()
```

### K-Mean-Clustring on `iris` Dataser

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

# -----------------------------------------
# 1. Load IRIS dataset
# -----------------------------------------
iris = load_iris()
X = iris.data[:, :2]   # only 2 features for visualization
feature_names = iris.feature_names[:2]

# -----------------------------------------
# 2. Apply K-Means
# -----------------------------------------
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

labels = kmeans.labels_                  # cluster each point belongs to
centroids = kmeans.cluster_centers_      # centroid positions

# -----------------------------------------
# 3. Visualization
# -----------------------------------------
plt.figure(figsize=(8, 6))

# scatter plot of points colored by cluster
plt.scatter(X[:, 0], X[:, 1], c=labels, s=80)

# scatter centroids
plt.scatter(
    centroids[:, 0], centroids[:, 1],
    marker='X', s=300, c='red', label="Centroids"
)

plt.title("K-Means Clustering on IRIS Dataset (Using 2 Features)")
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.legend()
plt.grid(True)
plt.show()
```
