```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 150000, 185000, 205000, 230000],  # Chai Point
])

print("==============ZOmato Sale Analysis====================")
print("Shape of sales data:",sales_data.shape)
print("Data of 1st 3 Restaurant: \n",sales_data[0:3])
print("Sales data with rnage:",sales_data[:,3:])
```

```python
# TOtal Sales Per Year
print(np.sum(sales_data, axis=0))
yearly_sales = np.sum(sales_data[:,1:],axis=0)
print("yearly sales:",yearly_sales)

# Minimum Sales per Restaurant
min_sales = np.min(sales_data[:,1:],axis=1)
print("Minimum sales of every restaurant:",min_sales)

#Maximum Sales per year
max_sale_per_year = np.max(sales_data[:,1:],axis=0)
print("Maximum Sales Per Year:",max_sale_per_year)

#Average Sales Per Restaurant
avg_sales_per_rest = np.average(sales_data[:,1:],axis=1)
print("Average Sales Per Restaurant:",avg_sales_per_rest)

#Cummulative Sale Restaurant
cum_sales_per_rest = np.cumsum(sales_data[:,1:],axis=1)  #Add Prev With current to give resultant
print("Cummulative Sales Per Rest(one restaurant's sales over multiple years):\n",cum_sales_per_rest)

#Average Cumulitive Sale
plt.Figure(figsize=(10,6))
plt.plot(np.average(cum_sales_per_rest,axis=0))
plt.title("Average cumulative sales for each year")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

## Vector Operations

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
print("Vector Addition:",vector1+vector2)
print("Vector Multiplication:",vector1*vector2)
print("Dot Product:",np.dot(vector1,vector2))
angle = np.arccos((np.dot(vector1,vector2)) / (np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print(angle)
```
