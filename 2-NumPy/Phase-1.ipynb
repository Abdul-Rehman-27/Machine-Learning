### Numpy Array vs List

```python

import time

py_list = [1,2,3]       #List multiplication repeats the list
print("List Multiplication", py_list*2)

np_array = np.array([1,2,3])      #NumPy array multiplication multiplies each element.
print("Array Multiplication", np_array*2)

start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List Operation time:",time.time()-start)

start = time.time()
np_array = np.arange(1000000)*2
print("\n Numpy Operation time:",time.time()-start)
```

## Creating Array from Scratch 

```python
zeros = np.zeros((3,4))
print("Zero Array: \n", zeros)

ones = np.ones((3,4))
print("One Array: \n", ones)

full = np.full((3,4),27)
print("Full Array: \n", full)

#We used this most 
random = np.random.random((3,4))
print("Random Array: \n", random)

#Random Array with Integer from 1 to 100
random = np.random.randint(0, 100, (3, 4))
print("Random Array of Integer:\n", random)

#Sequence Array 
sequence = np.arange(0,10,2)
print("Sequence Array:\n",sequence)

```

## Vector, Matrix and Tenser

```python
#Vector(1D Array)
vector = np .array([1,2,3])
print("Vector:",vector)

#Matrix(2D Array)
matrix = np.array([[1,2,3],[4,5,6]])
print("Matrix:\n",matrix)

#Tensor(Multidimensional Array)
tensor = np.array([
  [[1,2],[3,4]],
  [[5,6],[7,8]]
  ])
print("Tensor:\n",tensor)
```

## Array Properties

```python
array = np.array([[1,2,3],
[4,5,6]
])
print("Shape:",array.shape)
print("Dimension:",array.ndim)
print("Size:",array.size)
print("Data Type:",array.dtype)
```

## Array Reshaping

```python
array = np.arange(12)
print("Array:",array)

reshaped = array.reshape((3,4))
print("Reshaped Array: \n",reshaped)

#Return copy of Array 
flatened = array.flatten()        #Convert to 1D
print("Flatened Array:",flatened)

#Return orignal Array instead of copy 
raveled = array.ravel()       
print("Raveled Array:",raveled)

#Transpose
transpose = reshaped.transpose()
print("Transponse:\n",transpose)
```
