## Visualizing Sales Data Using Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt0
```

```python
df = pd.read_csv("netflix.csv")
df
```

```python
df.columns
```

```python
df = df.dropna(subset = ['show_id','type','title','director','cast','country','date_added','release_year','rating','duration','listed_in','description'])
```

```python
type_count = df['type'].value_counts()
print(type_count)
plt.bar(type_count.index,type_count.values,color=['red','blue'],width=0.5, label='2025 Product Sales')
plt.xlabel('Types')
plt.ylabel('Counts')
plt.title('Movies vs TV Shows')
plt.tight_layout()
plt.savefig('movies_vs_tvshows_bar.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
rate_counts = df['rating'].value_counts()
print(rate_counts)
plt.pie(rate_counts,labels=rate_counts.index,autopct='%1.1f%%',colors=['Purple','Orange','Blue','Red'])
plt.title('Percentage of Content Rating')
plt.savefig('rating_counts_pie.png', dpi=300, bbox_inches='tight')   # Done this before show
plt.show()
```

```python
# Step 2: Filter only movies
movies = df[df['type'] == 'Movie']

# Step 3: Remove 'min' from duration and convert to numeric
movies['duration_int'] = movies['duration'].str.replace(' min', '').astype(float)

# Step 4: Plot histogram
plt.hist(movies['duration_int'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Movie Durations on Netflix')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.savefig('Movies_Duration_Hist.png', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
release_counts = df['release_year'].value_counts().sort_index()
print(release_counts)
plt.figure(figsize=(10, 6))
plt.scatter(release_counts.index, release_counts.values, color='blue')
plt.title('Release Year vs Number No of Shows')
plt.xlabel('Release Year')
plt.ylabel('Number of Shows')
plt.tight_layout()
plt.savefig('relase_year_scatter.png')
plt.show()
```
