## Understanding Some key terms

```python
#Key Terms 
'''
Data Point:-
A data point is simply a single piece of data plotted on a graph. It represents a pair of values — usually an x-coordinate and a y-coordinate — that you want to display on a chart. (x,y)

X-axis & Y-axis:-
1. X-axis (Horizontal Line → left to right)
2. Y-axis (Vertical Line ↑ up and down)

Figure:-
In Matplotlib, a Figure is the overall window or canvas where everything is drawn — like the whole page that can contain one or more plots.

Axes:-
An Axes in Matplotlib is the actual plot area inside the Figure where data is visualized.
It includes the x-axis, y-axis, and the graph itself (such as lines, bars, or points).
Each Axes represents a single graph.

Plot:-
The visual representation of data (like a line, bar, scatter, etc.) drawn on a graph. It is the final output we actuallt see.

Marker:-
A marker is a symbol (like a dot, circle, square, etc.) that shows the exact position of each data point on a plot.

Line:-
The actual curve or straight path that connects data points on a plot.

Line Style:-
The appearance or pattern of the line — whether it's solid, dashed, dotted, etc.

Color:-
Color defines the visual color of the line, marker, or bar in the plot.

Legend:-
The legend shows a small box on the graph explaining what each line or plot represents (based on labels).

Label:-
A label is the name given to a line or plot, used for identifying it in the legend.

Title:-
A title is the heading of the graph that explains what the entire plot is about.

Grid:- 
Background lines that help you match points with the x and y axes.

Function:- 
A function is a block of code that does a specific task, and you can use it again and again by calling its name. eg.plot()

Method
A method is a function that belongs to an object and is called using that object.

Parameters:-
Parameters are the names defined in a function that accept values when you call the function.

Arguments
Arguments are the actual values you pass to the function when you call it.

Objec oriented API:-
In Matplotlib, the Object-Oriented (OO) API is a way to create plots by working directly with figure and axes objects, instead of using simple pyplot functions.
      Why Use OOP Style?
         1-More control over complex layouts
         2-Better for subplots, multiple figures, and customized plots
         3-Cleaner code for larger project
         4-Pyplot (uses global state): It automatically creates a figure and axes behind the scenes.
         5-You manually create a figure and an axes. Then you use the axes object (ax) to draw things.

You don’t need to write code for creating them.

DPI:-
DPI stands for Dots Per Inch. It measures the resolution or quality of a saved image or figure.
    1-Higher DPI means more dots per inch → sharper and clearer images.
    2-Lower DPI means fewer dots per inch → images may look blurry or pixelated.

Backend:-
A backend in Matplotlib is the engine that handles rendering and displaying plots.
It controls how and where the plots are drawn — whether on your screen, saved to a file, or shown in a notebook.
Examples: 'TkAgg', 'Qt5Agg', 'GTK3Agg'
'''

```
