## Understanding Pyplot

```python
# What is Pyplot
'''
Pyplot is a module in the Matplotlib library used for creating plots and charts easily.
Think of it like:Just like a notebook and pen for drawing data visually, pyplot is a toolbox that helps draw graphs in Python.
Pyplot gives simple functions to draw lines, graphs, bars, etc. like plot(), title(), xlabel(), and show().
'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = ['Mod','Tue','Wed','Thu','Fri'] # x-axis
y = [10,20,15,18,7]
plt.plot(x,y)
plt.title('Bakery Sales This Weeek')
plt.xlabel('Days')
plt.ylabel('Sales Per Day')
plt.show()
```
