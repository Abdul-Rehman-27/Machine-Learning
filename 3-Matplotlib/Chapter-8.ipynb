## Subplots and layout adjustments

```python
import matplotlib.pyplot as plt
```

```python
# Why subplot?
'''
1-If u want to compare multiple charts side by side
2-Space Efficient
3-U can quickly anaylyse related datasets together
'''
```

```python
#plt.subplot(nrows,ncolumn, index)  #Note:- here will not be zero based
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1)  # 1 row, 2 Column, 1st subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2)  # 1 row, 2 Column, 2nd subplot
plt.bar(x,y)
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python
# Recommended Way
#fig, ax = plt.subplots(nrows, ncols, figsize=(width, height)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y)
ax[0].set_title('Line Plot')

ax[1].bar(x,y)
ax[1].set_title('Bar Chart')

plt.tight_layout()
plt.show()
```

## Individually Customizing Subplot


```python
#fig, ax = plt.subplots(nrows, ncols, figsize=(width, height)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y,color='red')
ax[0].set_title('Line Plot')

ax[1].bar(x,y,color='green')
ax[1].set_title('Bar Chart')

fig.suptitle('Comparison between Line Graph and Bar Chart') #Setting Title of Whole figure

plt.tight_layout()
plt.show()
```
