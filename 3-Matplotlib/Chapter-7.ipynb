## Scatter Plot

```python
# Used to find the correlation b\w two variables (Hours vs Study)
```

```python
import matplotlib.pyplot as plt
```

```python
#plt.scatter(x,y,color='colorname',marker='marker style',labe='label_name')
hours = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [40, 45, 50, 55, 60, 65, 70, 75]
plt.scatter(hours, marks, color='red', marker='^', label='Student Data')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')
plt.title('Study Hours vs Marks Scatter Plot')
plt.legend()
plt.grid(True)
plt.show()

# Result:- More Study Directly Proportional to Highest Marks
```

```python
# Comparing Two Data Sets
plt.scatter([1,2,3], [50,55,60], color='red', label='Class A') #Group A
plt.scatter([1,2,3], [45,50,52], color='blue', label='Class B') #Group B
plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')
plt.title('Comparison Of Two Classes')
plt.legend()
plt.grid(True)
plt.show()

#Result:- Student of Class A is more Intelligent than Class B
```
