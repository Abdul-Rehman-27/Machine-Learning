## Saving Figures

```python
import matplotlib.pyplot as plt
```

```python
# Savefig('filename with extension', dpi=value, bbox_inches='tight')

#plt.pie(values,lables=label_list,colors=color_list,autpct='%1.1f%%') #1.1 mean 1 digit before decimanl 1 digits after decimal 
reigons = ['North','South','East','West']
revenue = [3000,2000,1500,1000]
plt.pie(revenue,labels=reigons,autopct='%1.1f%%',colors=['Purple','Orange','Blue','Red'])
plt.title('Revenue Contribution By Reigon')
plt.savefig('Pie-Chart.png', dpi=300, bbox_inches='tight')   # Done this before show
plt.show()
```
