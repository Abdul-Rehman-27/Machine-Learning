## Plotting bar charts,pie charts and histogram 

```python
import matplotlib.pyplot as plt
```

## Bar Charts

```python
#plt.bar(x,hieght,color='color name',width=value, label='label name')
products = ['PA','PB','PC','PD']
sales = [1000,1500,800,1200]
plt.bar(products,sales,color='gray',width=0.5, label='2025 Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Prouct Sales Comparison')
plt.xticks( ['PA','PB','PC','PD'],['Product A','Product B','Product C','Product D'])
plt.legend()
plt.show()
```

```python
# If u want to draw Horizontally
#plt.bar(x,hieght,color='color name',width=value, label='label name')
products = ['PA','PB','PC','PD']
sales = [1000,1500,800,1200]
plt.barh(products,sales,color='gray',label='2025 Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Prouct Sales Comparison')
plt.legend()
plt.show()
```

## Pie Charts

```python
#plt.pie(values,lables=label_list,colors=color_list,autpct='%1.1f%%') #1.1 mean 1 digit before decimanl 1 digits after decimal 
reigons = ['North','South','East','West']
revenue = [3000,2000,1500,1000]
plt.pie(revenue,labels=reigons,autopct='%1.1f%%',colors=['Purple','Orange','Blue','Red'])
plt.title('Revenue Contribution By Reigon')
plt.show()
```

## Histogram

```python
#plt.hist(data, bins=number of bins, color='colorname', egecolor='black')
marks = [45, 48, 50, 52, 58, 60, 61, 65, 70, 72, 75, 80, 85, 90]
plt.hist(marks, bins=5, color='skyblue', edgecolor='black', label='Student Marks')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.title('Marks Distribution of Students')
plt.legend()
plt.show()
```
