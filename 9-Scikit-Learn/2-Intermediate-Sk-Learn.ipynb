## Intermediate Ml with sk learn 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1.Simple Linear Regression 

```python
from sklearn.linear_model import LinearRegression

#load the dataset
df = sns.load_dataset('tips')

#select the feature Column from df
X = df[['total_bill']]

#select the Label from df
y = df['tip']

#load the linear regression model
model = LinearRegression()

#fit the model means train the model
model.fit(X,y)

#let Predicts the tip for an unknown bill
model.predict([[45]])
```

```python
sns.lmplot(x='total_bill', y='tip', data=df)
#Below given line in scatter plot is known as best fit line
plt.show()
```

```python
# what is best fit line?
# The best fit line is a straight line that best represents the data points in a scattier plot

# what is gradient descent?
# Gradient descent is a method computers use to find the best line by making small changes to the line, step by step, to make the errors smaller each time.
```

```python
from sklearn.linear_model import LinearRegression

#load the dataset
df = sns.load_dataset('tips')

#select the feature Column from df
X = df[['total_bill','size']]

#select the Label from df
y = df['tip']

#load the linear regression model
model = LinearRegression()

#fit the model means train the model
model.fit(X,y)

#let Predicts the tip for an unknown bill
input_data = pd.DataFrame([[45, 2]], columns=['total_bill', 'size']) #Inorder to remove warning
print(model.predict(input_data))
```

## Encoding

```python
from sklearn.linear_model import LinearRegression

#load the dataset
df = sns.load_dataset('tips')

#encoding smoker(Manually)
# df['smoker'] = df['smoker'].map({'Yes': 1, 'No': 0})
# df.head()

#Encoding Smoker
df['num_smoker'] = df['smoker'].map({'Yes': 1, 'No': 0})
df.head()

#select the feature Column from df
X = df[['total_bill','size','num_smoker']]

#select the Label from df
y = df['tip']

#load the linear regression model
model = LinearRegression()

#fit the model means train the model
model.fit(X,y)

#let Predicts the tip for an unknown bill
input_data = pd.DataFrame([[5.527, 2, 1]], columns=['total_bill', 'size', 'num_smoker']) #Inorder to remove warning
print(model.predict(input_data))
```

# 3.Logistic Regressoion (Classification) and below is Binary Classification

```python
from sklearn.linear_model import LogisticRegression

#load the dataset
df = sns.load_dataset('tips')

#select the feature Column from df
X = df[['total_bill','size','tip']]

#select the Label from df
y = df['smoker']

#load the linear regression model
model = LogisticRegression()

#fit the model means train the model
model.fit(X,y)

#let Predicts the tip for an unknown bill
input_data = pd.DataFrame([[5.527, 2, 1]], columns=['total_bill', 'size', 'tip']) #Inorder to remove warning
print(model.predict(input_data))

#Why Warning?
#The warning happens because the model was trained with column names but the prediction was made using a list without column names, which can confuse the model about feature order.
```

## 4. Naive Bayes

```python
# from sklearn.naive_bayes import GaussianNB #With Alias

from sklearn.naive_bayes import GaussianNB

#load the dataset
df = sns.load_dataset('tips')

#select the feature Column from df 
X = df[['total_bill','size','tip']]

#select the Label from df
y = df['smoker']

#load the linear regression model
model = GaussianNB()

#fit the model means train the model
model.fit(X,y)

#let Predicts the tip for an unknown bill
input_data = pd.DataFrame([[34, 3, 4]], columns=['total_bill', 'size', 'tip']) #Inorder to remove warning
print(model.predict(input_data))
```

## 5.Decision Tree (Classifier)

```python
from sklearn.tree import DecisionTreeClassifier

#load the dataset
df = sns.load_dataset('tips')

#select the feature Column from df 
X = df[['total_bill','size','tip']]

#select the Label from df
y = df['smoker']

#load the linear regression model
model = DecisionTreeClassifier()

#fit the model means train the model
model.fit(X,y)

#let Predicts the tip for an unknown bill
input_data = pd.DataFrame([[50, 2, 5]], columns=['total_bill', 'size', 'tip']) #Inorder to remove warning
print(model.predict(input_data))
```

## 6.Decision Tree (Regressor)

```python
from sklearn.tree import DecisionTreeRegressor

#load the dataset
df = sns.load_dataset('tips')

#select the feature Column from df 
X = df[['total_bill','size',]]

#select the Label from df
y = df['tip']

#load the linear regression model
model = DecisionTreeRegressor()

#fit the model means train the model
model.fit(X,y)

#let Predicts the tip for an unknown bill
input_data = pd.DataFrame([[50, 2]], columns=['total_bill', 'size']) #Inorder to remove warning
print(model.predict(input_data))
```
