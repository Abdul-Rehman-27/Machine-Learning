```python
from sklearn.neighbors import KNeighborsClassifier

X = [
    [180,7],
    [200,7.5],
    [160,6],
    [240,8],
    [220,7.8],
    [300,9]
]

# 0=Apple , 1=Orange
Y = [0,0,0,1,1,1]

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,Y)

weight = float(input("Enter weight in grams: "))
size = float(input("Enter size in cm: "))

result = model.predict([[weight,size]])[0]
print(result)
if result == 0:
    print("Fruit is Apple")
else:
    print("Fruit is Orange")


```
